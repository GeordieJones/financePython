```python
from tqdm import tqdm
from datetime import datetime
import yfinance as yf

latest_known = datetime(2024, 9, 30)

# Load tickers from your DataFrame
df_ticker = pd.read_pickle('year0_data.pkl')
tickers = df_ticker.index.tolist()  # Use .index, not 'symbol', since 'symbol' is the index in your dataset

tickers_with_recent_quarters = []

for ticker in tqdm(tickers, desc="Checking quarterly financials"):
    try:
        q_fin = yf.Ticker(ticker).quarterly_financials
        if q_fin is not None and not q_fin.empty:
            if not q_fin.empty and q_fin.columns[0] > latest_known:
                tickers_with_recent_quarters.append(ticker)
    except Exception as e:
        print(f"Failed on {ticker}: {e}")

# Print result
print(f"\nTickers with available recent quarterly financials ({len(tickers_with_recent_quarters)}):")
print(tickers_with_recent_quarters)
```
