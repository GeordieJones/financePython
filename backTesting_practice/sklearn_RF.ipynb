```python
import numpy
print(numpy.__version__)

import pandas as pd
print("Pandas version:", pd.__version__)

import sklearn
print("sklearn version:", sklearn.__version__)


data = pd.read_pickle('data.pkl')

data.head()
```

```python
data
```

```python
ticker_list = data.index.get_level_values('ticker').unique().tolist()
ticker_list
```

```python
import yfinance as yf
end_date = '2023-10-27'
start_date = '2021-10-31'

df = yf.download(tickers=ticker_list, start = start_date, end = end_date)
```

```python
df = df['Close']
df
```

```python
df_returns = df - df.iloc[0]
df_returns
```

```python
all_indicators_df = data[['garman_klass_vol', 'rsi', 'bb_low', 'bb_mid', 'bb_high', 'atr', 'macd', 'Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
farma_french_df = data[['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']]
vol_indicators_df = data[['garman_klass_vol', 'rsi', 'bb_low', 'bb_mid', 'bb_high', 'atr', 'macd']]
farma_french_df

```

```python
vol_indicators_df
```

```python
'''
farma_french_df 2021-10-31 to 2023-09-30
df_returns 2021-11-01 to 2023-10-26

'''
lag_days = 21
X_list = []
y_list = []
for ticker in ticker_list:
    X_ticker = farma_french_df.xs(ticker, level='ticker')
    y_ticker = df_returns[ticker].shift(-lag_days)
    common_dates = X_ticker.index.intersection(y_ticker.index)

    X_ticker = X_ticker.loc[common_dates]
    y_ticker = y_ticker.loc[common_dates]
    combined = pd.concat([X_ticker, y_ticker], axis=1).dropna()

    X_ticker = combined[X_ticker.columns]
    y_ticker = combined[y_ticker.name]

    X_ticker = X_ticker.copy()
    X_ticker['ticker'] = ticker

    X_list.append(X_ticker)
    y_list.append(y_ticker)

X_all = pd.concat(X_list)
y_all = pd.concat(y_list)
X_all = X_all.reset_index()
y_all = y_all.reset_index(drop=True)

y_all
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#X_all = X_all.drop(columns=['index'])

categorical_features = ['ticker']
numeric_features = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ],
    remainder='passthrough'
)
X_preprocessed = preprocessor.fit_transform(X_all)
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y_all, test_size=0.2, random_state=42, shuffle=False)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2:", r2_score(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(y_test.values, label='Actual Returns', alpha=0.7)
plt.plot(y_pred, label='Predicted Returns', alpha=0.7)
plt.title('Random Forest: Actual vs Predicted Returns')
plt.xlabel('Sample Index')
plt.ylabel('Returns')
plt.legend()
plt.show()
```

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.hlines(0, y_pred.min(), y_pred.max(), colors='r', linestyles='dashed')
plt.xlabel('Predicted Returns')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()
```

Frech Farma numbers on there own do not give a ML model enough of a indicator to be able to predict the returns in one month. As shown in the graph this is mostly do to the shear volatility of the market which is partially trimmed by the regressor which leaves the market to be guessed as random

```python
# Same as with the french farma but now with volatility metrics

lag_days = 21
X_list = []
y_list = []
for ticker in ticker_list:
    X_ticker = vol_indicators_df.xs(ticker, level='ticker')
    y_ticker = df_returns[ticker].shift(-lag_days)
    common_dates = X_ticker.index.intersection(y_ticker.index)

    X_ticker = X_ticker.loc[common_dates]
    y_ticker = y_ticker.loc[common_dates]
    combined = pd.concat([X_ticker, y_ticker], axis=1).dropna()

    X_ticker = combined[X_ticker.columns]
    y_ticker = combined[y_ticker.name]

    X_ticker = X_ticker.copy()
    X_ticker['ticker'] = ticker

    X_list.append(X_ticker)
    y_list.append(y_ticker)

X_all = pd.concat(X_list)
y_all = pd.concat(y_list)
X_all = X_all.reset_index()
y_all = y_all.reset_index(drop=True)

y_all

```

```python
X_all = X_all.drop(columns=['index'])
X_all
```

```python
def ML(X,y):
    categorical_features = ['ticker']

    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ],
        remainder='passthrough'
    )
    X_preprocessed = preprocessor.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42, shuffle=False)

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print("MSE:", mean_squared_error(y_test, y_pred))
    print("R2:", r2_score(y_test, y_pred))
    return model, y_test, y_pred
```

this model using volatility seems to look much better the only issue is in its predictions durning down markets. As seen in the first graph during normal and positive markets it seems acurate but durning down term it actually predicts the wrong way. This is most likely due to a combination of two issues. 1: not normilizing the data by leaving it in different scales it slows down the learning process for the limited data. 2: just using volitlity as growth the predictions always seem to go up when volitlity goes up.
